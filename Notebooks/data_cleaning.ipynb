imputation des valeurs manquantes 

```python
cat_data_train = []
num_data_train = []

for i , c in enumerate(df_train.dtypes):
    if c == object:
        cat_data_train.append(df_train.iloc[:, i])
    else:
        num_data_train.append(df_train.iloc[:, i])
cat_data_train = pd.DataFrame(cat_data_train).transpose()
num_data_train = pd.DataFrame(num_data_train).transpose()

#cat_data_train
#num_data_train
```

```python
#df_train.iloc[:, 4]
```

1. Variables catégorielles 

```python
# pour les variables catégorielles on va remplacer les valeurs manquantes par la valeur la plus fréquente

cat_data_train = cat_data_train.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data_train.isnull().sum().sort_values(ascending=False).to_frame()
#cat_data_train.isnull().sum().any()
```

```python
cat_data_train['Education'].value_counts()
```

2. Variables numeriques

```python
num_data_train.fillna(method='bfill', inplace=True)
num_data_train.isnull().sum().sort_values(ascending=False).to_frame()
```

```python
# variable cible
#df_train['Loan_Status'].value_counts()
target_value = {'Y': 1, 'N': 0}
target = cat_data_train['Loan_Status']
cat_data_train.drop('Loan_Status', axis=1, inplace=True)
target = target.map(target_value)
target
```

```python
# recodage des modalités des variables catégorielles
le = LabelEncoder()
for i in cat_data_train:
    cat_data_train[i] = le.fit_transform(cat_data_train[i])
    
    
cat_data_train.head()
```
